#### Use BeutifulSoup to convert html to dataframe.

```python
soup = BeautifulSoup(html2, 'html.parser')
table = soup.find("table")

# The first tr contains the field names.
headings = [th.get_text().strip() for th in table.find("tr").find_all("th")]


datasets = []
for row in table.find_all("tr")[1:]:
    dataset = dict(zip(headings, (td.get_text() for td in row.find_all("td"))))
    datasets.append(dataset)


df=pd.DataFrame(datasets)
df.head(20)
```

### Remove last two unwanted characters

```python
df['Borough'] = df['Borough'].map(lambda x: str(x)[:-1])
df['Postal code'] = df['Postal code'].map(lambda x: str(x)[:-1])
df['Neighborhood'] = df['Neighborhood'].map(lambda x: str(x)[:-1])
df.head(30)
```

### Change column order

```python
#change column order
df = df[["Postal code", "Borough", "Neighborhood"]] 
df.head()
```

### Rename column

```python
#rename column
df.rename(columns = {'Postal code':'PostalCode'}, inplace = True) 
df.head()
```

```python
#filter rows with Borough not assigned
print(df.shape)
df = df[df.Borough != "Not assigned"]
print(df.shape)
df = df.reset_index(drop=True)
print(df.shape)
df.head()
```
